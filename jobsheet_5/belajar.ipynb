```python
class Komputer:
  def __init__(self, merk, processor, ram_gb):
    self.merk = merk
    self.processor = processor
    self.ram_gb = ram_gb

  def info_spesifikasi(self):
    print("=" * 60)
    print(f"Menampilkan info dari {self.__class__.__name__}")
    print(f"Merk: {self.merk}")
    print(f"Processor: {self.processor}")
    print(f"Ram: {self.ram_gb} GB")

  def jalankan_aplikasi(self, nama_aplikasi):
    print("=" * 60)
    print(f"{self.__class__.__name__} {self.merk} menjalankan aplikasi: {nama_aplikasi}")

  def install_software(self, *args):
    print("=" * 60)
    print(f"Menginstall software di {self.__class__.__name__} {self.merk}")
    
    if not args:
      print("->Tidak ada software yang diinstall")
      return 0
    
    for i in args:
      print(f"->Software {i} terinstall")

class Laptop(Komputer):
  def __init__(self, merk, processor, ram_gb, ukuran_layar_inch, berat_kg):
    super().__init__(merk, processor, ram_gb)
    self.ukuran_layar_inch = ukuran_layar_inch
    self.berat_kg = berat_kg
  
  def info_spesifikasi(self):
    super().info_spesifikasi()
    print(f"Ukuran layar: {self.ukuran_layar_inch} inch")
    print(f"Berat: {self.berat_kg} Kg")

class Desktop(Komputer):
  def __init__(self, merk, processor, ram_gb, jenis_casing, monitor_external):
    super().__init__(merk, processor, ram_gb)
    self.jenis_casing = jenis_casing
    self.monitor_external = monitor_external
  
  def info_spesifikasi(self):
    super().info_spesifikasi()
    print(f"Jenis casing: {self.jenis_casing}")
    print(f"Monitor External: {self.monitor_external}")


def cetak_semua_spesifikasi(devices):
  print("=" * 60)
  print("Menampilkan semua info spesifikasi dari device yang tersedia")
  for device in devices:
    device.info_spesifikasi()
  
if __name__ == "__main__":
  komputer1 = Komputer("Lenovo", "Intel Core I5 10th Gen",16)
  laptop1 = Laptop("HP", "AMD Ryzen 3 3250P",8 , 10, 0.8)
  desktop1 = Desktop("Asus", "Intel Core I3 11th Gen",8 , "Tower", True)
  devices = [komputer1, laptop1, desktop1]

  print("=" * 60)
  print("DEMONSTRASI POLIMORFISME")
  cetak_semua_spesifikasi(devices)
  print("=" * 60)
  print()

  print("=" * 60)
  print("Simulasi Overloading")
  komputer1.install_software("Corel Draw")
  komputer1.install_software("Photoshop", "Figma")
  komputer1.install_software()
  print("=" * 60)
```
